```
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import random


gdrive_path = '/content/gdrive/My Drive/Colab Notebooks/'

cooked_path = gdrive_path + 'Project_GogiMaster/datasets/cooked/'
uncooked_path = gdrive_path + 'Project_GogiMaster/datasets/uncooked/'
cooked_list = os.listdir(cooked_path)
uncooked_list = os.listdir(uncooked_path)
# print(cooked_list)
# breakpoint()

gogi_to_np = list()
y_label = list()


for image_name in uncooked_list:
    img = cv2.imread(uncooked_path + image_name)
    print(img.shape)

    #   uncooked image 이상치 제거  #
    if img.mean() > 240:
      continue

    plt.subplot(131)
    plt.imshow(img)
    # plt.show()
    # print('Mean value :', img.mean())
    # breakpoint()

    gogi_to_np.append(img)
    y_label.append(0)
# breakpoint()

for image_name in cooked_list:
    img = cv2.imread(cooked_path + image_name)
    print(img.shape)

    #   cooked image 이상치 제거  #
    if img.mean() > 200:
      continue

    plt.subplot(131)
    plt.imshow(img)
    # plt.show()    
    # print('Mean value :', img.mean())
    # breakpoint()

    gogi_to_np.append(img)
    y_label.append(1)

# breakpoint
seed_number = 326
random.seed(seed_number)
random.shuffle(gogi_to_np)
random.seed(seed_number)
random.shuffle(y_label)
print(y_label)

np.save(gdrive_path + 'Project_GogiMaster/datasets/' + 'gogi_to_np', np.array(gogi_to_np))
np.save(gdrive_path + 'Project_GogiMaster/datasets/' + 'y_label', np.array(y_label))
```
